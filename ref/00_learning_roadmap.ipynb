## 学习方法论

### 每个主题的学习流程

1. **读理论** → 理解算法的数学推导和直觉
2. **从零实现** → 手写核心代码，深入理解算法
3. **Sklearn 实现** → 使用工业级库快速验证
4. **可视化** → 直观理解模型行为
5. **调参与实验** → 修改参数观察影响
6. **完成练习** → 巩固理解

### 推荐资源
