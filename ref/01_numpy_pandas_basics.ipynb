# NumPy & Pandas 基础回顾

本 Notebook 回顾机器学习中必备的 NumPy 和 Pandas 核心操作。

## 1. NumPy 核心操作

```python
import numpy as np
print(f"NumPy 版本: {np.__version__}")
```

### 1.1 创建数组

```python
arr_list = np.array([1, 2, 3, 4, 5])
print(f"从列表创建: {arr_list}")

arr_zeros = np.zeros((3, 4))
print(f"全零数组 (3x4):\n{arr_zeros}")

arr_ones = np.ones((2, 3))
print(f"全一数组 (2x3):\n{arr_ones}")

arr_arange = np.arange(0, 10, 2)
print(f"arange: {arr_arange}")

arr_linspace = np.linspace(0, 1, 5)
print(f"linspace: {arr_linspace}")

arr_random = np.random.randn(3, 3)
print(f"随机正态分布 (3x3):\n{arr_random}")

arr_randint = np.random.randint(0, 100, (2, 5))
print(f"随机整数 (2x5):\n{arr_randint}")
```

### 1.2 数组属性与基本运算

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])
print(f"形状: {arr.shape}")
print(f"维度: {arr.ndim}")
print(f"元素类型: {arr.dtype}")
print(f"总元素数: {arr.size}")

print(f"\narr + 10 =\n{arr + 10}")
print(f"arr * 2 =\n{arr * 2}")
print(f"arr ** 2 =\n{arr ** 2}")
print(f"arr 转置:\n{arr.T}")
print(f"点积 arr @ arr.T:\n{arr @ arr.T}")
```

### 1.3 索引与切片

```python
arr = np.arange(0, 24).reshape(4, 6)
print(f"原始数组 (4x6):\n{arr}\n")
print(f"第2行: {arr[1]}")
print(f"最后一列: {arr[:, -1]}")
print(f"前2行, 中间3列:\n{arr[:2, 1:4]}")
print(f"布尔索引 (>10):\n{arr[arr > 10]}")
```

### 1.4 常用数学函数

```python
arr = np.array([1, 4, 9, 16, 25])
print(f"sqrt: {np.sqrt(arr)}")
print(f"log: {np.log(arr)}")
print(f"exp: {np.exp(np.array([1, 2, 3]))}")
print(f"sin: {np.sin(np.array([0, np.pi/2, np.pi]))}")
print(f"均值: {arr.mean()}, 标准差: {arr.std()}")
print(f"最大值: {arr.max()}, 最小值: {arr.min()}, 求和: {arr.sum()}")
print(f"沿axis=0的均值: {np.array([[1,2],[3,4]]).mean(axis=0)}")
```

## 2. Pandas 核心操作

```python
import pandas as pd
print(f"Pandas 版本: {pd.__version__}")
```

### 2.1 创建 DataFrame

```python
df = pd.DataFrame({
    '姓名': ['张三', '李四', '王五', '赵六'],
    '年龄': [25, 30, 22, 28],
    '城市': ['北京', '上海', '广州', '深圳'],
    '薪资': [15000, 20000, 12000, 18000]
})
df
```

### 2.2 数据查看与筛选

```python
print(f"前2行:\n{df.head(2)}\n")
print(f"列信息:")
print(df.info())
print(f"\n描述统计:\n{df.describe()}")
print(f"\n薪资 > 15000:\n{df[df['薪资'] > 15000]}")
print(f"\n按薪资排序:\n{df.sort_values('薪资', ascending=False)}")
```

### 2.3 缺失值处理

```python
df_na = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8],
    'C': [9, 10, 11, 12]
})
print(f"原始数据:\n{df_na}\n")
print(f"缺失值统计:\n{df_na.isnull().sum()}\n")
print(f"删除含NaN的行:\n{df_na.dropna()}\n")
print(f"填充NaN为均值:\n{df_na.fillna(df_na.mean())}")
```

### 2.4 GroupBy 与聚合

```python
df_sales = pd.DataFrame({
    '产品': ['A', 'B', 'A', 'B', 'A', 'C'],
    '区域': ['北', '北', '南', '南', '北', '北'],
    '销量': [100, 200, 150, 180, 120, 90],
    '价格': [10, 20, 10, 20, 10, 30]
})
print(f"按产品分组求均值:\n{df_sales.groupby('产品').mean(numeric_only=True)}")
print(f"\n按产品和区域分组求和:\n{df_sales.groupby(['产品', '区域']).sum(numeric_only=True)}")
print(f"\n多聚合函数:\n{df_sales.groupby('产品').agg({'销量': ['sum', 'mean'], '价格': 'mean'})}")
```

### 2.5 合并与连接

```python
employees = pd.DataFrame({'emp_id': [1, 2, 3], 'name': ['张三', '李四', '王五']})
salaries = pd.DataFrame({'emp_id': [1, 2, 4], 'salary': [15000, 20000, 12000]})

print(f"Inner Join:\n{employees.merge(salaries, on='emp_id', how='inner')}\n")
print(f"Left Join:\n{employees.merge(salaries, on='emp_id', how='left')}\n")
print(f"Outer Join:\n{employees.merge(salaries, on='emp_id', how='outer')}")
```
